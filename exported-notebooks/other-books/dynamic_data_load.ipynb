```python
import os
import pandas as pd

import warnings

warnings.filterwarnings('ignore', category=pd.errors.SettingWithCopyWarning)
warnings.filterwarnings('ignore', category=FutureWarning)
```

```python
FOLDER_PATH = 'C:/Users/Dhyey2.Patel/Desktop/Dhyey2.Patel/PVC_import_analysis_&_forecasting/data/raw-data'
master_df = pd.DataFrame()
```

```python
for file_name in os.listdir(FOLDER_PATH):

    if file_name.endswith('.xlsx') or file_name.endswith('.xls'):

        file_path = os.path.join(FOLDER_PATH, file_name)
        
        print(f'Processing : {file_name}')

        temp_df = pd.read_excel(file_path)
        temp_df.columns = temp_df.columns.str.strip().str.lower()
        temp_df = temp_df[['date','product', 'k level', 'fin qty (mts)', 'fin cif ($/mt)']]

        master_df = pd.concat([master_df, temp_df], ignore_index=True)

        print(f'{file_name} successfully appended')
```

```python
master_df['date'] = pd.to_datetime(master_df['date'], format='mixed')
master_df.sort_values(by='date', inplace=True)
master_df['product'].fillna('pvc-susp', inplace=True)
master_df = master_df[~master_df['product'].isin(['PVC Paste', 'c-PVC'])]
master_df.dropna(inplace=True)
print('Final Aggregrated Dataset: ')
master_df
```
